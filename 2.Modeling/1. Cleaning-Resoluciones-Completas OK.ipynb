```python
#librerias

#generales
import pandas as pd
from pprint import pprint
import numpy as np

#para leer docx

import docx
import exception
```

```python
## Importar data
```

```python
#funcion para leer data

def leerExcel(año):
    dataAño = pd.read_excel("Data-resoluciones.xlsx",
                             sheet_name=año,
                             usecols=[0,1,2],
                             header=None)
    return dataAño
```

```python
#subiendo datos de cada año y concatenando --> MEJORAR

Data1erQuinquenio = pd.concat([leerExcel("2006"), leerExcel("2007"), 
                       leerExcel("2008"), leerExcel("2009"),
                       leerExcel("2010")], 
                      ignore_index=True)
```

```python
## Convertir en diccionario
```

```python
#funcion para convertir a dic de dic

def DicDeDic(data):

    filas = len(data)
    variables = list( data.iloc[:,0].value_counts().index )

    resoluciones = {}
    num_res = 0

    for fila in range(filas):
        # Cada resolución será un diccionario dentro de un diccionario más grande

        # Debe crearse una resolución si es la primera fila o si la última fila revisada fue "Decisión"
        if fila == 0 or data.iloc[fila-1,0] == 'Decisión':
            num_res += 1
            resoluciones[num_res] = {}

        # Si en la primera columna hay una variable, creamos una llave en el diccionario de la resolución
        if data.iloc[fila, 0] in variables:
            variable = data.iloc[fila, 0]
            resoluciones[num_res][variable] = ""

        #En las otras dos columnas, si el valor no es un vacío se agregará a la última variable registrada como llave
        cant_colums = 2
        
        for columna in range(1, cant_colums):
            celda = data.iloc[fila, columna]
            if pd.notna( celda ):
                # Si el valor de la variable llave es vacío, no necesita agregarle espacios
                if resoluciones[num_res][variable] == "":
                    resoluciones[num_res][variable] = str(celda)
                else:
                    resoluciones[num_res][variable] += " " + str(celda)

    return resoluciones
```

```python
#dic de dic para cada año

Dic1erQuinquenio = DicDeDic(Data1erQuinquenio)
```

```python
pprint(Dic1erQuinquenio)
```

```python
# convertir en DF
```

```python
#En dataframe el diccionario general

def DicADF(DicDeDic):
    resultado = pd.DataFrame()
    for resolucion in DicDeDic:
        temp = pd.DataFrame.from_dict(DicDeDic[resolucion], orient='index')
        resultado = pd.concat( [resultado, temp], axis=1 )

    resultado = resultado.transpose()
    resultado.index = list(range(1, len(resultado.index)+1) )
    
    return resultado
```

```python
#convirtiendo a DF el dic creado anteriormente
DF1erQuinquenio = DicADF(Dic1erQuinquenio)
DF1erQuinquenio
```

```python
## Agregar criterior aplicables
```

```python
#añadiendo informacion de word: Criterios aplicables a cada resolución

### pip3 install python-docx ####

def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return fullText
```

```python
#trayendo todo el word en una lista
doc = getText('criterios-elias.docx')

#convirtiendolo en numpy array para trabajarlo mejor
numpyDoc = np.array(doc)
```

```python
#borrando filas donde hay espacios vacios
np.delete(numpyDoc,np.where(numpyDoc == ''))
np.delete(numpyDoc,np.where(numpyDoc == ' '))
```

```python
#creando diccionarios con id_resol y textos
id_resol = list(DF1erQuinquenio["Número de resolución"])
resolus = {}

for row in range(len(numpyDoc)):
# Cada id_resolucion será clave de un diccionario que contendra todos los párrafos

    # Debe crearse una resolución si es la primera fila o 
    # si la última fila revisada está dentro de la lista de id_resol
    
    if row == 0 or numpyDoc[row][:] in id_resol:
        codigo = numpyDoc[row][:]
        resolus[codigo]=''
        
    else:          
        resolus[codigo]+=numpyDoc[row][:]
```

```python
#convirtiendo el anterior dic con los parrafos de word a un DF
parrafosResol = pd.DataFrame.from_dict(resolus, orient = "index")

#reiniciando index para evitar problemas
parrafosResol=parrafosResol.reset_index()
```

```python
#renombrando
parrafosResol = parrafosResol.rename(columns={"index": "Número de resolución", 0: "Criterios aplicables"})
```

```python
#haciendo merge entre parrafos de word y DF de todas las resoluciones

DFTOTAL1erQuinquenio = pd.merge(parrafosResol, DF1erQuinquenio, on='Número de resolución')
DFTOTAL1erQuinquenio
```

```python
## borrar Criterios aplicables_y y setear correctmente la columna

DFTOTAL1erQuinquenio = DFTOTAL1erQuinquenio.drop(['Criterios aplicables_y'], axis=1)
DFTOTAL1erQuinquenio
```

```python
#rename columna Criterios

DFTOTAL1erQuinquenio = DFTOTAL1erQuinquenio.rename(columns={"Criterios aplicables_x": "Criterios aplicables"})
```
